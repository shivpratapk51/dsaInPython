```python
def twoSum(nums: list, target: int):
    # using the hashmap
    map = {}
    for idx, i in enumerate(nums):
        comp = target - i
        if comp in map:
            return [map[comp], idx]
        map[i] = idx
    return []

num = [2,7,11,15]
target = 9
print(twoSum(num, target))
```

```python
def containsDuplicate(nums:list):
    hs = set()
    for i in nums:
        if i in hs:
            return True
        hs.add(i)
    return False
        

num1 = [2,14,18,22,22]
num2 = [1,2,3,4]
num3 = [1,1,1,3,3,4,3,2,4,2]

print("Num1: ", containsDuplicate(num1))
print("Num2: ", containsDuplicate(num2))
print("Num3: ", containsDuplicate(num3))
```

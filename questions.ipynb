```python
def twoSum(nums: list, target: int):
    # using the hashmap
    map = {}
    for idx, i in enumerate(nums):
        comp = target - i
        if comp in map:
            return [map[comp], idx]
        map[i] = idx
    return []

num = [2,7,11,15]
target = 9
print(twoSum(num, target))
```
